```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
df = sns.load_dataset('iris')

# Display the first few rows of the dataset
print(df.head())

# Get the basic information about the dataset
print(df.info())

# Summary statistics
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())

# Distribution of the target variable (species)
print(df['species'].value_counts())

# Plotting the distribution of each feature
df.hist(bins=20, figsize=(12, 8))
plt.suptitle("Distribution of Features")
plt.show()
```

```python
# Pairplot to visualize pairwise relationships and distributions
sns.pairplot(df, hue='species', markers=["o", "s", "D"])
plt.suptitle("Pairwise Relationships and Distributions", y=1.02)
plt.show()

# Scatter plot of Sepal Length vs Sepal Width, colored by species
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
# Compute the correlation matrix, excluding the 'species' column
corr_matrix = df.drop('species', axis=1).corr()

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Violin plot to show the distribution of features across species
plt.figure(figsize=(12, 8))
sns.violinplot(x='species', y='sepal_length', data=df, inner='quartile', palette='muted')
plt.title("Violin Plot of Sepal Length by Species")
plt.show()

plt.figure(figsize=(12, 8))
sns.violinplot(x='species', y='petal_length', data=df, inner='quartile', palette='muted')
plt.title("Violin Plot of Petal Length by Species")
plt.show()
```

```python
# KDE plot for Sepal Length
plt.figure(figsize=(8, 6))
sns.kdeplot(df['sepal_length'], shade=True, color='r')
plt.title("KDE Plot of Sepal Length")
plt.show()

# KDE plot of Sepal Length by Species
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='sepal_length', hue='species', fill=True, common_norm=False, palette='Set1')
plt.title("KDE Plot of Sepal Length by Species")
plt.show()
```

```python
# Pairwise scatter plots with regression lines
sns.pairplot(df, hue='species', markers=["o", "s", "D"], kind='reg')
plt.suptitle("Pairwise Scatter Plots with Regression Lines", y=1.02)
plt.show()
```

```python
# FacetGrid with histogram
g = sns.FacetGrid(df, col="species", height=5, aspect=1)
g.map(sns.histplot, "petal_length", kde=True, bins=15)
g.add_legend()
plt.suptitle("Histogram of Petal Length by Species", y=1.02)
plt.show()
```

```python
# Bar plot showing the average petal length for each species
plt.figure(figsize=(8, 6))
sns.barplot(x='species', y='petal_length', data=df, estimator=np.mean, ci='sd', palette='Set2')
plt.title("Average Petal Length by Species")
plt.show()
```
